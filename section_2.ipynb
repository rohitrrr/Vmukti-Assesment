# 1. Difference Between CNN and RNN

### **Convolutional Neural Network (CNN)**

CNNs are designed to process data with a grid-like topology, such as images. The architecture uses convolutional layers to extract spatial features like edges, textures, and patterns, enabling the model to understand spatial hierarchies in data. CNNs are particularly good at recognizing patterns in fixed-size inputs.
  * Convolutional layers apply filters to the input to detect local patterns.
  * Pooling layers reduce the spatial dimensions of the data to make computation efficient.
  * Fully connected layers transform the extracted features into output predictions.
  * CNNs work on fixed-sized inputs and focus on spatial hierarchies in data. They assume independence between different parts of the input (e.g., pixels in an image).
  * CNNs assume no temporal dependency between input data points. Each region in the input is treated independently of the others (except when spatially close).
  * CNNs process data in parallel, making them computationally efficient for tasks like image recognition, where each convolutional filter operates on all parts of the image simultaneously.
  * Image Classification: CNNs are ideal for tasks like classifying images into categories. For instance, a CNN could classify whether an image contains a cat or a dog by analyzing spatial patterns.


### **Recurrent Neural Network (RNN)**

RNNs are designed for sequential data, where the order of data points matters. They process inputs one at a time while maintaining a hidden state that stores contextual information about the sequence. RNNs excel at capturing temporal dependencies.
  * A recurrent layer processes input sequences step-by-step.
  * Hidden states maintain information from previous steps to inform future predictions.
  * Variants like LSTMs and GRUs address the problem of vanishing gradients by adding gates to regulate the flow of information.
  * RNNs handle variable-length sequences of data, where each element in the sequence is processed in relation to others. They are inherently sequential, making them suitable for tasks involving time-series data or language.
  * RNNs maintain dependencies across time steps. They use hidden states to capture relationships between earlier and later parts of a sequence.
  * RNNs process data sequentially, making them computationally intensive. Each time step depends on the output of the previous step, limiting parallelization.
  * Sentiment Analysis: RNNs are well-suited for tasks like determining the sentiment of a text (positive or negative). By analyzing the sequence of words, the RNN can capture the meaning and emotional tone of the text.





# 2. CNN Model Architecture

* The convolution layers extract features from an image (input). A small filter or kernel scans through the image and extracts features, for example, a verticle or a horizontal line, and creates a feature map. The layer that comes after the convolution layer is the pooling layer.
* The pooling layer essentially downsamples the feature map extracted by the convolution layer.
* A feature map extracted by the convolution layer contains the exact position of the feature.
* This may result in overfitting. The pooling layer runs a filter across the features map.
* It only takes the specific information from that filter, either the average of the values coming under the filter or the maximum depending on the pooling method selected. This reduces the feature map's spatial size and translates the feature's exact spatial information to rough information.
* This helps to prevent overfitting. Any convolution and pooling layers can be stacked together depending on the complexity of the input data set.
* The final pooling layer is then flattened out and transformed into a one-dimensional array, and fed to the fully connected layers that predict the output.
