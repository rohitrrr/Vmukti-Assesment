### 2 Explain how data augmentation techniques (e.g., cropping, flipping, rotation) help improve the performance of computer vision models.


Data augmentation techniques like cropping, flipping, and rotation improve the performance of computer vision models by:

1. Enhancing Diversity: Augmentation creates variations in the dataset, helping models generalize better to unseen data by exposing them to different perspectives, scales, and orientations of objects.

2. Reducing Overfitting: By artificially expanding the dataset, models are less likely to memorize training examples, leading to improved generalization during testing.

3. Simulating Real-world Scenarios: Techniques like rotation and flipping mimic real-world variations, enabling models to handle different viewpoints, lighting conditions, and noise effectively.

4. Increasing Robustness: Augmented data introduces challenges like occlusions and partial visibility, making the model more robust to similar situations in deployment.

5. Maximizing Limited Data: For datasets with limited quantity, augmentation effectively increases the training set size without requiring additional data collection.
